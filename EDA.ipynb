```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/ericpool/Documents/GitHub/sprint_4_project/vehicles_us.csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
plt.hist(df['condition'], bins=50, edgecolor='k')
plt.title('Car Condition')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python



plt.figure(figsize=(10, 6))
plt.hist(df['odometer'], bins=50, edgecolor='k')
plt.title('Car Odometer')
plt.xlabel('Odometer')
plt.ylabel('Price')
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['type'], df['price'], alpha=0.5)
plt.title('Car Type')
plt.xlabel('Type')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python

df['model_year'] = df.groupby('model')['model_year'].transform(lambda x: x.fillna(x.median()))

df['cylinders'] = df.groupby('model')['cylinders'].transform(lambda x: x.fillna(x.median()))

df['odometer'] = df.groupby(['model', 'model_year'])['odometer'].transform(lambda x: x.fillna(x.median()))

```
